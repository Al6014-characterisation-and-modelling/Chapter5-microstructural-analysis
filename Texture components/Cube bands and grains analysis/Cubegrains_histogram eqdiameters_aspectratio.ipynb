# Histograms plots

### Import packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import h5py
import pandas as pd
```

# 2D cube grains

Import files

```python
# Small maps data
#BA_ebsdRD="BA_NDRD/BA_RD_smallGrainsData.csv"
#NoBA_ebsdRD="NoBA_NDRD/NoBA_RD_smallGrainsData.csv"
#BA_ebsdTD="BA_NDTD/BA_TD_small_GrainsData.csv"
#NoBA_ebsdTD="NoBA_NDTD/NoBA_RD_smallGrainsData.csv"

# Large maps data
BA_ebsdRD="BA_NDRD/BA_RD_large_GrainsData.csv"
NoBA_ebsdRD="NoBA_NDRD/NoBA_RD_largeGrainsData.csv"
BA_ebsdTD="BA_NDTD/BA_TD_large_GrainsData.csv"
NoBA_ebsdTD="NoBA_NDTD/NoBA_TD_largeGrainsData.csv"
```

Introduce the name for the exported file

```python
output_filename_2DEBSD_RD = 'cube_hist_NDRD'
output_filename_2DEBSD_TD = 'cube_hist_NDTD'
output_filename_2DEBSD_IA = 'cube_hist_IA'
output_filename_2DEBSD_NoIA = 'cube_hist_NoIA'
```

Read data from csv files and save it into new variables

```python
BARD = pd.read_csv(BA_ebsdRD)
NoBARD = pd.read_csv(NoBA_ebsdRD)
BATD = pd.read_csv(BA_ebsdTD)
NoBATD = pd.read_csv(NoBA_ebsdTD)

eqdiameters_BARD = BARD['EquivalentDiameter']
eqdiameters_NoBARD = NoBARD['EquivalentDiameter']
eqdiameters_BATD = BATD['EquivalentDiameter']
eqdiameters_NoBATD = NoBATD['EquivalentDiameter']

aspectratio_BARD = BARD['AspectRatio']
aspectratio_NoBARD = NoBARD['AspectRatio']
aspectratio_BATD = BATD['AspectRatio']
aspectratio_NoBATD = NoBATD['AspectRatio']

ylim = 600 #for small maps, use 80 ylim
```

Plot the histograms

```python
num_grains_BARD = len(eqdiameters_BARD)
num_grains_NoBARD = len(eqdiameters_NoBARD)

label_BARD = f'IA grains - NDRD slice \n (number of grains={num_grains_BARD})'
label_NoBARD = f'No IA grains - NDRD slice \n (number of grains={num_grains_NoBARD})'

plt.figure(figsize=(10, 6))
sns.histplot(eqdiameters_BARD, color='black', binwidth=5, kde=False, label=label_BARD, alpha=0.7,edgecolor='black', linewidth=1.2)
sns.histplot(eqdiameters_NoBARD,binwidth=5, kde=False, label=label_NoBARD, alpha=0.7,edgecolor='darkblue', linewidth=1.2)

plt.xlim(0, 90)
plt.ylim(0,ylim)

x_ticks = np.arange(0, 90, 5)
plt.xticks(x_ticks, fontsize=16)  

plt.xlabel('Equivalent diameter (µm)', fontsize=22)
plt.ylabel('Number of grains', fontsize=22)
#plt.title('Grain size comparison', fontsize=22)

plt.legend(fontsize=16)

plt.savefig(output_filename_2DEBSD_RD + '_eqdiameter.png', dpi=300, bbox_inches='tight')

plt.grid(True)
plt.show()
```

### Aspect ratio vs eq diameter

```python
bin_edges = np.arange(0, 95, 5)

df_BARD = pd.DataFrame({'Equivalent Diameter': eqdiameters_BARD, 'Aspect Ratio': aspectratio_BARD})
df_BARD['Diameter Bin'] = pd.cut(df_BARD['Equivalent Diameter'], bins=bin_edges)
avg_BARD = df_BARD.groupby('Diameter Bin').mean()

df_NoBARD = pd.DataFrame({'Equivalent Diameter': eqdiameters_NoBARD, 'Aspect Ratio': aspectratio_NoBARD})
df_NoBARD['Diameter Bin'] = pd.cut(df_NoBARD['Equivalent Diameter'], bins=bin_edges)
avg_NoBARD = df_NoBARD.groupby('Diameter Bin').mean()

bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

plt.figure(figsize=(10, 6))

plt.scatter(eqdiameters_BARD, aspectratio_BARD, color='black', alpha=0.8, label='IA grains - NDRD slice')
plt.scatter(eqdiameters_NoBARD, aspectratio_NoBARD,color='darkblue',alpha=0.5, label='No IA grains - NDRD slice')

plt.plot(bin_centers, avg_BARD['Aspect Ratio'], color='red', linestyle='-', linewidth=2, label='Avg IA grains')
plt.plot(bin_centers, avg_NoBARD['Aspect Ratio'], color='red', linestyle='--', linewidth=2, label='Avg No IA grains')

plt.xlabel('Equivalent Diameter (µm)', fontsize=22)
plt.ylabel('Aspect Ratio', fontsize=22)

y_ticks = np.arange(0, 25, 1.5)
plt.yticks(y_ticks, fontsize=16)  
plt.xlim(0, 90)
plt.legend(fontsize=16)

plt.grid(True, linestyle='--', alpha=0.6)

plt.savefig('aspect_ratio_vs_eqdiameterRD.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Aspect Ratio', data=df_box, palette='Set2')

plt.xlabel('') 
plt.ylabel('Aspect Ratio', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

plt.savefig('cube_boxplot_aspect_ratio.png', dpi=300)
plt.show()
```
