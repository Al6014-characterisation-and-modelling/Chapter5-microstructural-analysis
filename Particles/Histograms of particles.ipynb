```python
# Import necessary libraries
import os
import pandas as pd
import glob
import numpy as np

# Specify the folder containing the CSV files
folder_path = './Particles_surface_and_averall_CSV_files_new'

# Use glob to get all CSV files in the specified folder
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

# Loop through all the CSV files
for file_path in csv_files:
    # Extract the file name from the path
    file_name = os.path.basename(file_path)
    
    # Generate the variable name using the first 10 characters of the file name
    # Convert the variable name to lowercase to ensure consistency
    var_name = file_name[:14].lower()
    
    # Read the first column of the CSV file
    df = pd.read_csv(file_path, encoding='ISO-8859-1', usecols=[0])
    
    # Dynamically assign the first column to a variable named after the file, in lowercase
    globals()[var_name] = df.iloc[:, 0]
    
    # Optionally, you can print the variable name to verify
    print(f"Variable '{var_name}' created with data from the first column of {file_name}")
```
