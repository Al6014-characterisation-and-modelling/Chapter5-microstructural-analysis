```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import glob
import numpy as np
```

```python
folder_path = './Particles_surface_and_averall_CSV_files_new'

csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

for file_path in csv_files:
    file_name = os.path.basename(file_path)
    var_name = file_name[:14].lower()
    df = pd.read_csv(file_path, encoding='ISO-8859-1', usecols=[0])
    globals()[var_name] = df.iloc[:, 0]
    print(f"Variable '{var_name}' created with data from the first column of {file_name}")
```
